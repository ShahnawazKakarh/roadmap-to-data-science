Task 3: Difference between pd.concat, pd.merge and pd.join

pd.concat() – Combine DataFrames Along Rows or Columns

```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'Omar']
})

df2 = pd.DataFrame({
    'ID': [4, 5],
    'Name': ['Zara', 'Hassan']
})

# Combine vertically (default: axis=0)
result = pd.concat([df1, df2])
print(result)
```

pd.merge() – Combine DataFrames Based on a Common Column (SQL-style JOIN)

```python
employees = pd.DataFrame({
    'EmpID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'Omar']
})

salaries = pd.DataFrame({
    'EmpID': [1, 2, 4],
    'Salary': [5000, 6000, 7000]
})

# Merge on common column 'EmpID'
result = pd.merge(employees, salaries, on='EmpID', how='inner')
print(result)
```

pd.join() – Join on Index or Key Columns

```python
df1 = pd.DataFrame({'Name': ['Ali', 'Sara', 'Omar']}, index=[1, 2, 3])
df2 = pd.DataFrame({'Salary': [5000, 6000, 7000]}, index=[1, 2, 4])

# Join by index
result = df1.join(df2, how='inner')
print(result)
```

# Deal Numeric Data

How can we deal with the numeric data -> ffill(), bfill(), mean(), median()

Forward Fill - ffill() - Fills missing values using the previous non-null value (value of index 5 is copied to index of 6)

```python
data.head(40)
```

```python
# ffill = forward fill - will update the value with one record earlier (upside value copy)

# data['street_num'] = data['street_num'].fillna(method='ffill')

data['street_num'].fillna(method='ffill', inplace=True)
data.isnull().sum()
```

```python
# ffill = forward fill - will update the value with one record earlier (upside value copy)

data['street_suffix'].fillna(method='ffill', inplace=True)
data['street_name'].fillna(method='ffill', inplace=True)

data.isnull().sum()
```

```python
data.head(40)
```

Backward Fill - bfill() - Fills missing values using the next non-null value (value of index 6 is copied to index of 5)

```python
data['city'].fillna(method='bfill', inplace=True)
data['state'].fillna(method='bfill', inplace=True)
data['postcode'].fillna(method='bfill', inplace=True)

data.head(40)
```

```python
data.isnull().sum()
```
